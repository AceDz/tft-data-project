```python
import pandas as pd

df = pd.read_json("../data/tft_matches.json")

god_stats = df.groupby("god").agg({"placement": "mean", "player": "count"}).sort_values("placement")

god_stats 
```

```python

df_traits = df[["player", "traits"]]
df_traits = df_traits.rename(columns = {"traits": "trait"})

df_traits = df_traits.explode("trait")

df_traits.merge(df, on ="player") \
    .groupby("trait")["placement"] \
    .mean() \
    .sort_values()

df_traits
```

```python

df_units = df[["player", "units"]]
df_units = df_units.explode("units")
df_units = df_units.rename(columns = {"units": "unit"})

df_units.merge(df, on="player") \
    .groupby("unit")["placement"] \
    .mean() \
    .sort_values()

df_units
```

```python

df_fact = df[["match_id", "player","placement","level"]]

df_traits = df[["player", "traits"]]
df_traits = df_traits.rename(columns = {"traits": "trait"})

df_units = df[["player", "units"]]
df_units = df_units.rename(columns = {"units": "unit"})

traits_full = df_traits.merge(df_fact, on="player").explode("trait")
traits_full = traits_full.groupby("trait")["placement"].mean().sort_values()
traits_full
```

```python
df_fact = df[["match_id", "player","placement","level"]].copy()

df_traits = df[["player", "traits"]].explode("traits")
df_traits = df_traits.rename(columns = {"traits": "trait"})
df_traits

df_units = df[["player", "units"]].explode("units")
df_units = df_units.rename(columns = {"units": "unit"}) 
df_units

fact_traits = df_fact.merge(df_traits, on="player", how="left")
fact_traits

fact_units = df_fact.merge(df_units, on="player", how="left")
fact_units

fact_traits.groupby("trait")["placement"].mean().sort_values()
fact_units.groupby("unit")["placement"].mean().sort_values()
df_fact.groupby("player")["placement"].mean().sort_values()
```
